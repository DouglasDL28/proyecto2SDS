## Split
Debido a la alta cantidad de datos que se tienen y el poco poder computacional con el que se cuenta se reducira las 8,921,483 a que sea 1/8 para poder manejarlo en el pre procesamiento de igual manera con el test set

```python
malware_data_train = pd.read_csv('train.csv', dtype=db_types)
malware_data_test = pd.read_csv('test.csv', dtype=db_types)
```

```python
count = int(malware_data_test.shape[0]/16)
new_test = malware_data_test.sample(n = count, random_state = 12)
new_test
```

```python
count = int(malware_data_train.shape[0]/16)
grouped = malware_data_train.groupby('HasDetections', group_keys=False)
new_train = grouped.apply(lambda x: x.sample(min(len(x), count), random_state = 12))
new_train
```

```python
new_train.groupby('HasDetections', group_keys=False).count()
```

```python
new_train.to_csv('new_train.csv')  
```
